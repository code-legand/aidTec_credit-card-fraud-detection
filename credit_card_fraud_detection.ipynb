### Removing outliers and oversampling minority class

```python
rows = 6
cols = 5
fig, ax = plt.subplots(figsize = (16, 16), nrows=rows, ncols = cols)
rcount, ccount = 0, 0
for feature in features:
  sns.boxplot(data = data, x=data[feature], ax=ax[rcount, ccount])
  ccount+=1
  if ccount==5:
    ccount = 0
    rcount +=1
plt.show()
```

```python
for feature in features:
  q1 = data[feature].quantile(0.25)
  q3 = data[feature].quantile(0.75)
  iqr = q3-q1
  ul = q3+1.5*iqr
  ll = q1-1.5*iqr
  data = data.loc[(data['Class']==1) | (data[feature]>ll) & (data[feature]<ul)]
```

```python
data['Class'].value_counts()
```

```python
from sklearn.utils import resample
d0 = data.loc[data['Class']==0]
d1 = data.loc[data['Class']==1]
d1 = resample(d1, n_samples=110900)
data = pd.concat((d0, d1))
```

```python
data['Class'].value_counts()
```

```python
rows = 6
cols = 5
fig, ax = plt.subplots(figsize = (16, 16), nrows=rows, ncols = cols)
rcount, ccount = 0, 0
for feature in features:
  sns.boxplot(data = data, x=data[feature], ax=ax[rcount, ccount])
  ccount+=1
  if ccount==5:
    ccount = 0
    rcount +=1
plt.show()
```

```python
for feature in features:
  q1 = data[feature].quantile(0.25)
  q3 = data[feature].quantile(0.75)
  iqr = q3-q1
  ul = q3+1.5*iqr
  ll = q1-1.5*iqr
  data = data.loc[(data[feature]>ll) & (data[feature]<ul)]
```

```python
data['Class'].value_counts()
```

```python
from sklearn.utils import resample
d0 = data.loc[data['Class']==0]
d1 = data.loc[data['Class']==1]
d1 = resample(d1, n_samples=len(d0))
data = pd.concat((d0, d1))
```

```python
data['Class'].value_counts()
```

```python
data = data.drop(columns = ["V6", "V8", "V13"])
```

## Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns = ["Class"])
Y = data["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
x_train[:5]
```

## Predictive Analysis

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1-score: ", f1_score(y_test, y_pred))
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1-score: ", f1_score(y_test, y_pred))
```

```python
model = MLPClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1-score: ", f1_score(y_test, y_pred))
```
